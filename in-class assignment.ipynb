```
import nltk; nltk.download('wordnet')
from nltk.corpus import wordnet
synonyms = []

for syn in wordnet.synsets("play"): #change the word you want to lookup here
	for l in syn.lemmas():
		synonyms.append(l.name())

print(set(synonyms))
print("Here is our calculation of the number of synonyms ", len(synonyms))
```

We use the wordnet library to calculate synonyms.
The first print statment shows the output of the synonyms associated with the word of choice. In this case, the word play).
The second print statement calculates the legnth of the list and outputs it to the screen.
